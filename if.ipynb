```python
__author__ = "liujr"
__data__ = "2020-5-10 13:23"

if True:
    print("条件判断为真")
if False:
    print("False")
if 1:
    print("1")
if 2:
    print("2")
if 3:
    print("3")
    
if False:
    print("True")
elif True:
    print("elif")
else:
    print("else")
print("------")    
"""
比较运算符
==  判断两个值是否相等
>
<
>=
<=
!=

"""
if 1 == 2:
    print("相等")
    
a = [1,2,3]
b = a
c = [1,2,3]
print(a)
print(b)
print(c)

if a == b:
    print("a和b相等")
if a == c:
    print("a和c相等")
if b == c:
    print("b和c相等")
    
print(id(a))
print(id(b))
print(id(c))

if a is b:
    print("a和b相等")
if a is c:
    print("a和c相等")
if b is c:
    print("b和c相等")


if "ab" > "ac": # 字符没有串，因此字符串的比较，是单个比较每个字符，再汇总
    print("a大于b")
else:
    print("a小于b")
print("------")    
"""
in 
not in 
"""
print("a" in "ab")
print("a" not in "ab")
print("------")

"""
and
or
not
"""

# if 1 and 2 and 3
print(1 and 0)
print(1 or 0)
print(not 0)
print("------")
"""
身份运算
is
is not
"""

a = [1,2,3]
b = a
c = [1,2,3]
print(a is b)
print(a is not b)
```
